```python
def removeDuplicates(nums):
    list1 = []
    temp_list = []
    u_score = []

    for i in range(len(nums)):

        if len(temp_list) <= 1:
            temp_list.append(nums[i])
        else:
            u_score.append("_")

        if not (nums[i] == nums[i+1]):
            list1.append(temp_list)
            temp_list = []

    list1.append(u_score)
    return list1

nums = [1,1,1,2,2,2,3]
print(removeDuplicates(nums))
```
